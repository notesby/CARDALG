```python
import matplotlib.pyplot as plt
import gdal
import numpy as np
import os
```

```python
def readraster(file):
    dataSource = gdal.Open(file)
    band = dataSource.GetRasterBand(1)
    band = band.ReadAsArray()
    return(dataSource, band)
```

```python
path = "data"
tFiles = sorted(list(filter(lambda x: ".tif" in x,os.listdir(path))))
files = []
for file in tFiles:
    source,band = readraster(f"data/{file}")
    files.append([file,source,band])
```

```python
files[0][0],files[0][2].max()
```

```python
fig,ax = plt.subplots(len(files)//2+1,2,figsize=(30,100))
for i,file in enumerate(files):
    x = i //2
    y = i % 2
    ax[x][y].matshow(file[2])
    ax[x][y].set_title(file[0])
```
