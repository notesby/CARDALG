```python
def readraster(file):
    dataSource = gdal.Open(file)
    band = dataSource.GetRasterBand(1)
    band = band.ReadAsArray()
    return(dataSource, band)
```

```python
stateFiles1 = ["Actual_1994.tif","cbddist.tif","roaddist.tif","dda_2021_government_restricted.tif","den1991.tif","slope.tif"]
stateFiles2 = ["Actual_1999.tif","cbddist.tif","roaddist.tif","dda_2021_government_restricted.tif","den2001.tif","slope.tif"]
stateFiles = [stateFiles1,stateFiles2]
states = []
path = "data"
for i,files in enumerate(stateFiles):
    state = []
    for file in files:
        state.append([file,*readraster(f"{path}/{file}")])
    states.append([i,state])
```

```python
def getDomain(state):
    #state = states[0][1]
    noLayers = len(state)
    noRows,noCols = np.shape(state[0][2])
    domain = {}
    for i in range(noRows):
        for j in range(noCols):
            for k in range(noLayers):
                prop1= state[k][0]
                lv1 =  state[k][2][i][j]
                if prop1 not in domain:
                    domain[prop1] = {}
                for z in range(noLayers):
                    if z == k:
                        continue
                    prop2 = state[z][0]
                    lv2 = state[z][2][i][j]
                    if prop2 not in domain[prop1]:
                        domain[prop1][prop2] = {}
                    if lv1 not in domain[prop1][prop2]:
                        domain[prop1][prop2][lv1] = {"total":0}
                    if lv2 not in domain[prop1][prop2][lv1]:
                        domain[prop1][prop2][lv1][lv2] = 0
                    domain[prop1][prop2][lv1][lv2] += 1
                    domain[prop1][prop2][lv1]["total"] += 1
    return domain
```

```python
for state in states:
    state.append(getDomain(state[1]))
```
